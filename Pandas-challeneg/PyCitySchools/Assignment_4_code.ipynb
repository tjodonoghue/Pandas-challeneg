```python
#Dependencies 
import pandas as pd 
from pathlib import Path

school_data_to_load = Path(r"C:\Users\Gamer PC\Desktop\Pandas_Challenge\Pandas-challeneg\Starter_Code (5)\Starter_Code\PyCitySchools\Resources\schools_complete.csv")
student_data_to_load = Path(r"C:\Users\Gamer PC\Desktop\Pandas_Challenge\Pandas-challeneg\Starter_Code (5)\Starter_Code\PyCitySchools\Resources\students_complete.csv")

#Read in csvs 
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

#Merge data sets
complete_data = pd.merge(school_data, student_data, how = "left", on=["school_name", "school_name"])
complete_data.head()
```

```python
#Finding total number of unique schools 
unique_schools = complete_data["school_name"].nunique()
print("There are ", unique_schools, "unique schools.")
```

```python
#Finding total number of students

#Eliminating duplicate schools so we can use the size column
individual_schools = complete_data.drop_duplicates(subset = "school_name")
total_students = individual_schools["size"].sum()
print("There are ", total_students, "students")
```

```python
#finding the budget 
total_budget = individual_schools["budget"].sum()
print("The total budget for all schools is ", total_budget)
```

```python
#finding average math score using mean command 
average_math_score = complete_data["math_score"].mean()
 
print("The average math score is ", average_math_score)
```

```python
#find average reading score using same process 

average_reading_score = complete_data["reading_score"].mean() 

print("The average reading score is ", average_reading_score)
```

```python
#Finding how many students passed math 

#First finding total number who passed 
passed_math = complete_data[(complete_data["math_score"] >= 70)].count()["student_name"]

#Find percent passing
percent_passing_math = (passed_math / total_students) * 100

print(percent_passing_math, " percent of students are passing math.")
```

```python
#Finding how many students passed reading using the same method 

passed_reading = complete_data[(complete_data["reading_score"] >= 70)].count()["student_name"]
 
percent_passing_reading = (passed_reading / total_students) * 100

print(percent_passing_reading, " percent of students are passing reading.")
```

```python
total_passing = complete_data[(complete_data['math_score'] >= 70) & (complete_data['reading_score'] >= 70)].count()["student_name"]

percent_passing = (total_passing / total_students) * 100

print(percent_passing, "percent of students are passing.")
```

```python
#HERE I NEED TO CREATE A DATAFRAME TO HOLD ALL OF THE KEY SUMMARY STATISTICS FOR WHATS ALREADY BEEN DONE
```

```python
#Moving on to School Summary statistics, will start by grouping by schools 
grouped_school_data= complete_data.groupby(["school_name"])
```

```python
#Creating the list that will hold later statistics 
school_statistics = []
```

```python
#creating a loop to iterate through all school data
for school, data in grouped_school_data:
    school_size = data["size"].iloc[0]
    school_type = data['type'].iloc[0]
    school_budget = data['budget'].iloc[0]
    per_student_budget = school_budget/school_size 
    school_average_math = data['math_score'].mean()
    school_average_reading = data['reading_score'].mean()
    passing_math = (data[data['math_score'] >= 70].count()["student_name"] / school_size) * 100
    passing_reading = (data[data['reading_score'] >= 70].count()["student_name"] / school_size) * 100
    overall_passing = ((data[(data['math_score'] >= 70) & (data['reading_score'] >= 70)].count()['student_name']) / school_size) * 100
#appending the list that we created
    school_statistics.append({
    "School Name": school, 
    "School Type": school_type,
    "Total Students": school_size,
    "School Budget": school_budget,
    "Per Student Budget": per_student_budget,
    "Average Math Score": school_average_math,
    "Average Reading Score": school_average_reading,
    "Percent Passing Math": passing_math,
    "Percent Passing Reading": passing_reading,
    "Percent Passing Overall": overall_passing
})

#creating the dataframe 
school_statistics_df = pd.DataFrame(school_statistics)
school_statistics_df
```

```python
#creating a csv for the DF
file_path = r'C:\Users\Gamer PC\Desktop\school_statistics.csv'
school_statistics_df.to_csv(file_path, sep =',', index = False)
```
